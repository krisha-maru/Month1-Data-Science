# Week 3: NumPy and Pandas for Data Manipulation

This notebook contains the Week 3 hands-on exercises and client project tasks.


```python
import numpy as np
import pandas as pd
```

## 1. NumPy Array Operations


```python
arr = np.array([10, 20, 30, 40])

print('Original:', arr)
print('Add 5:', arr + 5)
print('Multiply by 2:', arr * 2)
print('Mean:', np.mean(arr))
```

## 2. NumPy Broadcasting


```python
arr = np.array([1, 2, 3])
result = arr + 10

print('Broadcasting result:', result)
```

## 3. Create a Pandas DataFrame


```python
data = {
    'Name': ['Krisha', 'Riya', 'Ankit', 'Megha'],
    'Marks': [85, 92, 78, 88]
}

df = pd.DataFrame(data)
print(df)
```

## 4. Data Filtering


```python
print(df[df['Marks'] > 85])
```

## 5. Handle Missing Values by Removing Rows


```python
data = {
    'Name': ['A', 'B', 'C', 'D'],
    'Marks': [85, np.nan, 78, 90]
}

df_missing = pd.DataFrame(data)
clean_df = df_missing.dropna()

print(clean_df)
```

## 6. Fill Missing Values with Mean


```python
df_filled = df_missing.copy()
df_filled['Marks'] = df_filled['Marks'].fillna(df_filled['Marks'].mean())
print(df_filled)
```

## 7. Grouping Data


```python
data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'Salary': [50000, 40000, 60000, 45000, 70000]
}

df_group = pd.DataFrame(data)
grouped = df_group.groupby('Department')['Salary'].mean()
print(grouped)
```

## 8. Full Data Cleaning Script


```python
data = {
    'Name': ['Krisha', 'Riya', 'Ankit', 'Krisha', None],
    'Marks': [85, 92, np.nan, 85, 78]
}

df_clean = pd.DataFrame(data)

# Remove duplicates
df_clean = df_clean.drop_duplicates()

# Fill missing marks with mean
df_clean['Marks'] = df_clean['Marks'].fillna(df_clean['Marks'].mean())

# Remove rows with missing names
df_clean = df_clean.dropna(subset=['Name'])

print('Cleaned Data:')
print(df_clean)
```

## 9. Aggregation Insights


```python
average_marks = df_clean['Marks'].mean()
max_marks = df_clean['Marks'].max()

print('Average Marks:', average_marks)
print('Max Marks:', max_marks)
```
